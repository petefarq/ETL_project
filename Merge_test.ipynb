```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file1 = "20190928-items.csv"
phones_df = pd.read_csv(csv_file1)
phones_df.head()
```

```python
phones_df.count()
```

```python
csv_file2 = "20190928-reviews.csv"
reviews_df = pd.read_csv(csv_file2)
reviews_df.count()
```

```python
merged_phone = pd.merge(reviews_df, phones_df, on="asin", how='left').fillna(value="Not Given")
merged_phone.head()
```

```python
merged_phone.count()
```

```python
merged_inner = pd.merge(reviews_df, phones_df, on="asin", how='inner').fillna(value="Not Given")
merged_inner.count()
```
