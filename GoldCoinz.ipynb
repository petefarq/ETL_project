```python
import pandas as pd
from sqlalchemy import create_engine
from config import rds_connection_string
import numpy as np
import excel
```

```python
csv_file1 = "20190928-items.csv"
reviews_df = pd.read_csv(csv_file1).dropna()
reviews_df.head()
```

```python
csv_file2 = "20190928-reviews.csv"
phones_df = pd.read_csv(csv_file2)
phones_df.head()
```

```python
merged_phone = pd.merge(reviews_df, phones_df, on="asin", how='left').dropna()
merged_phone.head()
```

```python
merged_phone.rename(columns={'title_x':'model', 'rating_x':'stars', 'title_y':'Review_Title'}, inplace=True)
merged_phone.head()
```

```python
del merged_phone['rating_y']
merged_phone.head()
```
